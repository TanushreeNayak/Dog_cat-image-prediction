Data Preprocessing

```python
import tensorflow as tf
# Instead of importing from keras directly, use tensorflow.keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
tf.__version__
```

```python
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
training_set = train_datagen.flow_from_directory(
        '/content/My_dataset/training_set/training_set',
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')
```

```python
!unzip '/content/My_dataset/test_set/test_set.zip' -d '/content/My_dataset/test_set'
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
        '/content/My_dataset/test_set/test_set',
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')
```

convolution

```python
cnn = tf.keras.models.Sequential()
```

pooling

```python
import keras
cnn = keras.models.Sequential()

# Create the Conv2D layer object first
conv2d_layer = keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3])

# Now add the layer object to the Sequential model
cnn.add(conv2d_layer) # Alternative syntax using original Keras imports
```

adding a second covolutional layer

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

flattening


```python
cnn.add(tf.keras.layers.Flatten())
```

full connection

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

output layer

```python
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

compilling the CNN

```python
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

Training the CNN

```python
cnn.fit(x = training_set, validation_data = test_set, epochs = 20)
```

Making a single prediction


```python
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('/content/My_dataset/test_set/test_set/cats/cat.4002.jpg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)
training_set.class_indices
if result[0][0] == 1:
  prediction = 'dog'
else:
  prediction = 'cat'
```
